```python
# General imports
import numpy as np
import pylatexenc

# Pre-defined ansatz circuit, operator class and visualization tools
from qiskit.circuit.library import QAOAAnsatz
from qiskit.quantum_info import SparsePauliOp
from qiskit.visualization import plot_distribution

# Qiskit Runtime
from qiskit_ibm_runtime import QiskitRuntimeService, Session
from qiskit_ibm_runtime import EstimatorV2 as Estimator
from qiskit_ibm_runtime import SamplerV2 as Sampler
from qiskit.primitives import StatevectorEstimator

# SciPy minimizer routine
from scipy.optimize import minimize

# rustworkx graph library
import rustworkx as rx
from rustworkx.visualization import mpl_draw
```

```python
num_nodes = 5
# The edge syntax is (start, end, weight)
edges = [(0, 1, 1), (0, 2, 1), (0, 3, 1), (0, 4, 1)]

G = rx.PyGraph()
G.add_nodes_from(range(num_nodes))
G.add_edges_from(edges)
mpl_draw(
    G, pos=rx.bipartite_layout(G, {0}), with_labels=True, node_color="#EE5396", font_color="#F4F4F4"
)
```

```python
# Problem to Hamiltonian operator
hamiltonian = SparsePauliOp.from_list([("IIIZZ", 1), ("IIZIZ", 1), ("IZIIZ", 1), ("ZIIIZ", 1)])
# QAOA ansatz circuit
ansatz = QAOAAnsatz(hamiltonian, reps=2)

#ansatz.decompose(reps=3).draw(output="mpl", style="iqp")
```

```python
# Optimization for real quantum hardware
# Optimizes for the native gates of the QC in question
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

target = backend.target
pm = generate_preset_pass_manager(target=target, optimization_level=3)

ansatz_isa = pm.run(ansatz)
#ansatz_isa.draw(output="mpl", idle_wires=False, style="iqp")

# Hamiltonian also has to be optimized
hamiltonian_isa = hamiltonian.apply_layout(ansatz_isa.layout)
hamiltonian_isa
```

download.png

```python
# The Estimator is what computes the expectation values of the ansatz through execution quickly, essentially our output layer
def cost_func(params, ansatz, hamiltonian, estimator):
    """Return estimate of energy from estimator

    Parameters:
        params (ndarray): Array of ansatz parameters
        ansatz (QuantumCircuit): Parameterized ansatz circuit
        hamiltonian (SparsePauliOp): Operator representation of Hamiltonian
        estimator (EstimatorV2): Estimator primitive instance

    Returns:
        float: Energy estimate
    """
    pub = (ansatz, [hamiltonian], [params])
    result = estimator.run(pubs=[pub]).result()
    cost = result[0].data.evs[0]

    return cost
```

```python
session = Session(backend=backend)

# Configure estimator
estimator = Estimator(session=session)
# estimator = StatevectorEstimator()
estimator.options.default_shots = 10_000
estimator.options.dynamical_decoupling.enable = True

# Configure sampler
sampler = Sampler(session=session)
sampler.options.default_shots = 10_000
sampler.options.dynamical_decoupling.enable = True
```

```python
# Initial Random parameters
x0 = 2 * np.pi * np.random.rand(ansatz_isa.num_parameters)
```

```python
# We use COBYLA for minimization as it is derivative free, on real quantum systems gradient methods usually perform better as the cost function "landscapes" are not barren
# we can test all optimizers that are implemented on IBM backends by simply swapping out the name here, or by creating our own custom method that we then invoke
from scipy.optimize import minimize

# Define your callback function
def callback(x):
    # Compute and print the cost function value at each step
    loss_value = cost_func(x, ansatz_isa, hamiltonian_isa, estimator)
    print(f"Loss at step: {loss_value}")

# Perform the optimization
res = minimize(cost_func, x0, args=(ansatz_isa, hamiltonian_isa, estimator), method="COBYLA", callback=callback)

# instead of minimize we can also define a stepwise custom optimizer, which come off the shelf in the github repo we'd looked at

#IBM doesn't really like it when you submit these many jobs so the notebook usually stalls here
#we can swap this out for the ionq token from fire instead
```

```python
# Result obtained in standard scipy output
res
```

```python
# Postprocessing
# Final parameters we have arrived at i.e creates the trained circuit
qc = ansatz.assign_parameters(res.x)
# Add measurements to our circuit
qc.measure_all()
qc_isa = pm.run(qc)
qc_isa.draw(output="mpl", idle_wires=False, style="iqp")

result = sampler.run([qc_isa]).result()
samp_dist = result[0].data.meas.get_counts()
# Close the session since we are now done with it
session.close()

# Since we only have 5 edges we can plot our results visually
# The most probable bit strings are our solutions, we should expect 2 as the cost function does not discriminate which nodes are assigned to which group
plot_distribution(samp_dist, figsize=(15, 5))


#If the notebook stalls, here's one I prepared earlier on an IonQ computer attached below
```
